```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
%matplotlib inline
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4,
                  cluster_std=1.8, random_state=101)
```

```python
data
```

```python
plt.scatter(x=data[0][:, 0], y=data[0][:, 1], c=data[1])
```

```python
from sklearn.cluster import KMeans
```

```python
kms = KMeans(n_clusters=4)
```

```python
kms.fit(data[0])
```

```python
kms.cluster_centers_
```

```python
kms.labels_
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10, 6))
ax1.set_title("K_means")
ax1.scatter(data[0][:, 0], data[0][:, 1],c=kms.labels_, cmap="copper")
ax2.set_title("Original")
ax2.scatter(data[0][:, 0], data[0][:, 1], c=data[1], cmap="copper")
```
